```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples=4,n_features=1,n_informative=1,noise=80,random_state=13)
```

```python
plt.scatter(X,y)
```

```python
# apply ols 

reg = LinearRegression()
reg.fit(X,y)

```

```python
reg.intercept_, reg.coef_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS line')
plt.show()
```

## now mathematically reaching to This line, intercept using Gradient descent

## Apply Gradient descent assuming slope is constant m = 78.35
## and lse's assume the starting value for intercept b =0

```python
y_pred = ((78.35*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS line')
plt.plot(X,y_pred,color='green',label='b=0 line')
plt.legend()
plt.show()
```

```python
m=78.35
b=0
loss_slope = -2 * np.sum(y-m*X.ravel()-b)
loss_slope
```

```python

# learning rate = 0.1
lr = 0.1
step_size = lr * loss_slope
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = ((78.35*X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS line')
plt.plot(X,y_pred,color='green',label='b=0 line')
plt.plot(X,y_pred1,color='pink',label='b=20.927763408209216')
plt.legend()
plt.show()
```
