```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/placement.csv')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

```

```python
X = df.iloc[:,0:1].values
y = df.iloc[:,-1].values
```

```python
X
```

```python
y
```

```python
plt.scatter(X, y)
plt.xlabel('CGPA')
plt.ylabel('Salary')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
X_test
```

```python
lr.predict(X_test)
```

```python
plt.scatter(X, y)
plt.plot(X_train, lr.predict(X_train), color='red')
plt.xlabel('CGPA')
plt.ylabel('Salary')
```

```python
m=lr.coef_  # slope
c=lr.intercept_  # y-intercept
```

```python
m,c
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
```

```python
y_pred =lr.predict(X_test)
```

```python
y_test
```

```python
print("Mean Absolute Error (in lpa): ", mean_absolute_error(y_test, y_pred))
```

```python
print("MSE : (in lpa_square)" , mean_squared_error(y_test, y_pred))
```

```python
print("RMSE : " ,np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
print("R2 score :", r2_score(y_test, y_pred))
```

## cgpa column explains 78% of variance in lpa

```python
## Adjusted R2 score = 1 - (1-R2)*(n-1)/(n-p-1)
r2 = r2_score(y_test, y_pred)
n = X_test.shape[0]
k = X_test.shape[1]
adj_r2 = 1 - (1-r2)*(n-1)/(n-k--1)
adj_r2
```

```python
## adding more features to see if we can improve the model
```

```python
new_df = df.copy()
```

```python
new_df['random_feature'] = np.random.random(200)
new_df=new_df[["cgpa","random_feature","package"]]
new_df.head()
```

```python
plt.scatter(new_df['random_feature'], new_df['package'])
plt.xlabel('random_feature')
plt.ylabel('package')
```

```python
X = new_df.iloc[:,0:2].values
y = new_df.iloc[:,-1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print("R2 Score : ", r2_score(y_test, y_pred))
```

## R2 score increased by addding a irrelevant colun

```python
print("Adjusted R2 Score : ", 1 - (1-r2_score(y_test, y_pred))*(X_test.shape[0]-1)/(X_test.shape[0]-X_test.shape[1]-1))
```

## but R2 adjusted decrease as expected due to irrelevant column
